```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a sample dataset
data = {
    'Age': [23, 25, 22, 30, 28, 35, 40, 23, 24, 22],
    'Height': [165, 170, 160, 175, 180, 172, 168, 158, 165, 170],
    'Weight': [65, 70, 60, 75, 80, 72, 68, 58, 65, 70],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male'],
    'Smoker': ['No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)
```

```python
# Step 1: Understand the Data
print("Data Types:\n", df.dtypes)
print("\nSummary Statistics:\n", df.describe())


```

```python
# Step 2: Univariate Analysis - Numerical Data
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)       #creates a subplot grid with 1 row and 2 columns,
sns.histplot(df['Age'], kde=True, color='blue')    #kernel density estimate (KDE), histplot- Seabornâ€™s histplot function
plt.title('Age Distribution (Discrete Data)')       #kde=True--The parameter kde=True adds a KDE curve on top of the histogram to show the estimated distribution of the data

plt.subplot(1, 2, 2)
sns.histplot(df['Height'], kde=True, color='green')
plt.title('Height Distribution (Continuous Data)')

plt.tight_layout()
plt.show()

```

```python
# Step 3: Univariate Analysis - Categorical Data
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.countplot(x='Gender', data=df, palette='Set1')    #  Gender wise ,  data source,  different color
plt.title('Gender Count (Categorical Data)')

plt.subplot(1, 2, 2)
sns.countplot(x='Smoker', data=df, palette='Set2')
plt.title('Smoker Count (Categorical Data)')

plt.tight_layout()
plt.show()

```

```python
# Step 4: Bivariate Analysis - Numerical Data
sns.scatterplot(x='Age', y='Weight', hue='Gender', data=df)    #plot by using different colors to represent different categories of the 'Gender' variable
plt.title('Scatter Plot of Age vs Weight')
plt.show()
```

```python
# Step 5: Outlier Detection - Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Height', data=df)
plt.title('Boxplot of Height by Gender')
plt.show()
```

```python
# Step 6: Feature Engineering- is the process of creating new features from existing data to improve the performance of a model or to gain additional insights
# Let's create a new feature: BMI (Body Mass Index)
df['BMI'] = df['Weight'] / (df['Height'] / 100) ** 2     # bmi=w/(h exp 2)
print("\nData with new BMI feature:\n", df)
```

```python

# Step 7: Visualization of the new feature
sns.histplot(df['BMI'], kde=True, color='purple')
plt.title('BMI Distribution')
plt.show()
```
